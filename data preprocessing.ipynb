```python

import pandas as pd
from pandas import DataFrame
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif

data = {
    'Name': ['hari','ravi','buvi','revi','sam'],
    'Age': [20,21,21,22,24],
    'Height': [120,121,122,123,127],
    'weight': [45,47,50,55,54],
    'Bp': [66,87,67,78,79]}

df = DataFrame(data)
print(df)
df.to_csv("data.csv", index=False)
df = pd.read_csv("data.csv")
df = df.replace(0, df.mean(numeric_only=True))
frame2 = df[['Age','Height','weight','Bp']]
scaler = StandardScaler()
scaled = scaler.fit_transform(frame2)
print(scaled)
print("mean:", scaled.mean(axis=0))
print("std deviation:", scaled.std(axis=0))
sel = SelectKBest(score_func=f_classif, k=2)
useful_features1 = sel.fit_transform(scaled, df['Age'])
print("Useful feature values based on StandardScaler")
print(useful_features1)
print(df)
```
